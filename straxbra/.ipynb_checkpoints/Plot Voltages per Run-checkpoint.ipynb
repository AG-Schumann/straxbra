```python
import utils
import matplotlib.pyplot as plt
import numpy as np
import straxbra
```

```python
run_docs = {}

for i in range(251):
    run_docs[i] = utils._GetRundoc(i)
```

```python
run_docs[250]
```

```python
kr_runs = {}
for k, v in run_docs.items():
    if v is None:
        continue
    if v['mode'] == 'kr-83m':
        kr_runs[k] = v
```

```python
x = np.array([k for k in kr_runs.keys() if k > 53])
an = np.array([v['anode_mean'] for k, v in kr_runs.items() if k > 53])
ca = np.array([v['cathode_mean'] for k, v in kr_runs.items() if k > 53])
ca_curr = np.array([v['cathode_current_mean'] for k, v in kr_runs.items() if k > 53])
ga = np.array([cv-280*cc for cv, cc in zip(ca, ca_curr)])
```

```python
plt.figure(figsize=(16, 9))
plt.plot(x, an, label='Anode', marker='o', c='b')
plt.plot(x, ca, label='Cathode', marker='o', c='orange')
plt.plot(x, ga, label='Gate', marker='o', c='green')
plt.xlabel('Magnitude Voltage [V]')
plt.ylabel('Run-ID')
plt.legend()
```

```python
plt.figure(figsize=(16, 9))
plt.title('Voltage-Difference in Region')
plt.plot(x, an+ga, label='amp', marker='o', c='red')
plt.plot(x, -ga+ca, label='drift', marker='o', c='k')
plt.ylim(2000, None)
plt.xlim(195, 205)
plt.legend()
```
