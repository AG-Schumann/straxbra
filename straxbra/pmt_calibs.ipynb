```python
gains = {}
runs = list(range(100, 251))

led_runs = \
    list(range(212, 217)) + \
    list(range(202, 207)) + \
    list(range(152, 157)) + \
    list(range(123, 131)) + \
    list(range(72, 77)) + \
    list(range(64, 68)) + \
    list(range(47, 51)) + \
    list(range(24, 33)) + \
    [165, 164, 97, 96, 61, 60, 43, 41, 35]

led_runs = [i for i in led_runs if (i <= max(runs) and i >= min(runs))]

for pmt in range(8):
    gains[pmt] = [utils.GetGains(run)[pmt] for run in runs]
```

```python
plt.figure(figsize=(16, 9))


for pmt, gain in gains.items():
    plt.plot(runs, gain, label=f'PMT{pmt}', marker='x')

for led_run in led_runs:
    plt.axvline(led_run, color='r', ls='-.')
```

```python
plt.figure(figsize=(16, 9))

colors = {
    0: 'blue',
    1: 'brown',
    2: 'pink',
    3: 'orange',
    4: 'purple',
    5: 'green',
    6: 'grey',
    7: 'red'
}

for pmt, gain in gains.items():
    plt.plot(runs, gain, label=f'PMT{pmt}', color=colors[pmt])


for led_run in led_runs:
    plt.axvline(led_run, color='r', ls='-.')


plt.xlim(200, 220)
plt.xticks(range(200, 221))
plt.legend(loc='best');
```
